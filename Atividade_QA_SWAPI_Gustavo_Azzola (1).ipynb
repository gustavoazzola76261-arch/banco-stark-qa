```python
import requests
import json

# -------------------------
# MODO DE SEGURANÇA (fallback)
# -------------------------
# Se a API não responder, o notebook usa dados locais
# para a atividade continuar funcionando.

DADOS_FALLBACK = {
    "https://swapi.dev/api/people/10/": {
        "name": "Obi-Wan Kenobi",
        "height": "182",
        "mass": "77",
        "eye_color": "blue-gray",
        "gender": "male",
        "url": "https://swapi.dev/api/people/10/"
    },
    "https://swapi.dev/api/people/13/": {
        "name": "Chewbacca",
        "height": "228",
        "mass": "112",
        "eye_color": "blue",
        "gender": "male",
        "url": "https://swapi.dev/api/people/13/"
    },
    "https://swapi.dev/api/planets/2/": {
        "name": "Alderaan",
        "climate": "temperate",
        "terrain": "grasslands, mountains",
        "population": "2000000000",
        "url": "https://swapi.dev/api/planets/2/"
    },
    "https://swapi.dev/api/starships/10/": {
        "name": "Millennium Falcon",
        "model": "YT-1300 light freighter",
        "manufacturer": "Corellian Engineering Corporation",
        "max_atmosphering_speed": "1050",
        "url": "https://swapi.dev/api/starships/10/"
    }
}

class RespostaLocal:
    def __init__(self, status_code, data):
        self.status_code = status_code
        self._data = data
        self.text = json.dumps(data, ensure_ascii=False)

    def json(self):
        return self._data

def buscar_endpoint(url):
    try:
        resposta = requests.get(url, timeout=10)
        if resposta.status_code == 200:
            return resposta, "online"
    except Exception:
        pass

    if url in DADOS_FALLBACK:
        return RespostaLocal(200, DADOS_FALLBACK[url]), "fallback local"

    return RespostaLocal(404, {"detail": "Not found"}), "fallback local"

print("Ambiente preparado com sucesso.")
```

### 3.2 Complete as validações

**Tarefa:** execute o bloco abaixo. Ele já está funcionando. Depois, alterem as expressões se o professor pedir outra nave ou outro endpoint. Em seguida, expliquem por que cada validação faz sentido para QA.


```python
url = "https://swapi.dev/api/starships/10/"
response, origem = buscar_endpoint(url)
dados = response.json()

print("Status 200:", response.status_code == 200)

# Os exemplos abaixo já estão corretos e executáveis.
# Se o professor pedir, vocês podem alterar para validar outros campos.
print("Campo 'name' existe:", "name" in dados)
print("Campo 'model' existe:", "model" in dados)
print("Nome não está vazio:", dados["name"] != "")
```

Resposta
1. Validação: response.status_code == 200
Por que faz sentido: É o primeiro portão. Se o status não for 200 (OK), significa que a API falhou (erro 500), o endereço está errado (erro 404) ou você não tem permissão (erro 403).

Visão de QA: Não adianta testar o nome da nave se a porta do servidor está fechada. É a validação de disponibilidade.

2. Validação: "name" in dados e "model" in dados
Por que faz sentido: Garante que a estrutura do JSON (o "contrato") não mudou.

Visão de QA: Imagine que o time de desenvolvimento mude a chave de name para ship_name. Se o seu front-end espera name, o site vai quebrar ou exibir "Undefined". Validar a existência da chave previne que quebras na interface cheguem ao usuário final.

3. Validação: dados["name"] != ""
Por que faz sentido: Verifica se o campo está populado. Um campo pode existir na estrutura, mas vir vazio ou nulo.

Visão de QA: De nada serve o campo existir se ele não traz a informação. Para o usuário, ver um card de nave sem nome é um erro de experiência (UX) tão grave quanto um erro de sistema.

## Parte 4 — Testes automatizados com assert

Agora vocês vão trabalhar sobre scripts já prontos. Os asserts abaixo já executam sem erro. A tarefa do grupo é **entender o que cada linha valida**, ajustar se o professor pedir outro recurso e registrar as respostas.


### 4.1 Assert no personagem

**Tarefa:** execute o bloco. Depois, expliquem o que cada assert valida e, se quiserem, alterem o endpoint para outro personagem.


```python
url = "https://swapi.dev/api/people/13/"
response, origem = buscar_endpoint(url)
dados = response.json()

assert response.status_code == 200, "Erro: status inesperado"
assert "name" in dados, "Erro: campo name não encontrado"
assert dados["name"] != "", "Erro: nome vazio"

print("Todos os testes do personagem passaram!")
```

respotas

Diferente do comando print, o comando assert é uma ferramenta fundamental de Testes Automatizados. Ele funciona como uma "sentinela": se a condição for verdadeira, o código segue em frente; se for falsa, o programa interrompe a execução na hora e exibe a mensagem de erro que você escreveu.

O que cada assert valida?
assert response.status_code == 200

O que valida: A comunicação com o servidor.

Por que é vital: Garante que a API está "viva" e respondeu com sucesso. Se o servidor cair ou o link estiver errado, o teste para aqui mesmo, evitando que você tente analisar dados que não existem.

assert "name" in dados

O que valida: A estrutura do contrato (Schema).

Por que é vital: Confirma que a API ainda entrega a chave name. Se os desenvolvedores da API mudarem o nome do campo para full_name, esse teste falha, avisando que o sistema que consome esses dados vai quebrar.

assert dados["name"] != ""

O que valida: A qualidade do conteúdo.

Por que é vital: Garante que o campo não está vindo como uma string vazia. De nada adianta a estrutura estar certa se a informação principal (o nome do personagem) estiver faltando.

### 4.2 Assert na nave

resposta

dados["model"] != ""

Campo: model (Modelo).

Objetivo: Validar se o conteúdo não está vazio. Confirma que, além de existir a chave, ela traz uma informação útil (ex: "YT-1300 light freighter").

## Parte 6 — Mini desafio final

Agora você vai alterar apenas o endpoint da função abaixo e gerar um pequeno resumo.

```python
def resumo_recurso(url):
    response, origem = buscar_endpoint(url)
    print("Origem dos dados:", origem)
    print("Status Code:", response.status_code)

    if response.status_code != 200:
        print("Recurso não encontrado.")
        return

    dados = response.json()

    for chave, valor in dados.items():
        print(f"{chave}: {valor}")

# ALTERE O ENDPOINT ABAIXO PARA TESTAR OUTRO RECURSO
resumo_recurso("https://swapi.dev/api/people/4/")
```
